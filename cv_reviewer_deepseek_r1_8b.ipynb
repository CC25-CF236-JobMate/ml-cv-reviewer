```python
from oyama import oyama
import ollama

model_path = "deepseek-r1:8b"
model_name = oyama.run(model_path)
```

```python
import pymupdf
import re

from google.colab import files
```

```python
def extract_text_from_pdf(fname):
  with pymupdf.open(fname) as doc:
    resume_text = chr(12).join([page.get_text() for page in doc])
    return resume_text
```

```python
def generate_response(text):
  response = ollama.chat(model=model_name, messages=[
    {
        'role': 'system',
        'content': """You are a professional HR with expertise in talent acquisition and CV evaluation. Your task is to review the provided CV and provide a detailed, constructive, and professional analysis. Focus on:
        - Clarity and structure of the CV
        - Relevance of skills and experience to a [insert job role, e.g., Software Engineer] position
        - Strengths and areas for improvement
        - Suggestions to enhance the CV
        - Job recommendation
        Respond in the CV original language, ensuring the tone is professional yet approachable. If specific job requirements or context are not provided, make reasonable assumptions based on common industry standards."""
    },
    {
      'role': 'user',
      'content': f"The CV: {text}",
    },
  ])
  return response.message.content
```

```python
def clean_response(response):
  pattern = r'<think>.*?</think>'
  return re.sub(pattern, '', response, flags=re.DOTALL)
```

```python
def main():
  f = files.upload()
  resume_text = ""
  for fn in f:
    resume_text = extract_text_from_pdf(fn)
  response = generate_response(resume_text)

  response = clean_response(response)
  print(response)
  return
```

```python
if __name__ == "__main__":
  main()
```
